We have pixel values, first of all we need to convert it into a list

```python
pixel_list = new_data['pixels'].to_list()
```

Let us create a numpy array, where each array represents a image, split using ''

```python
Arr = []
for pix in pixel_list:
    img = [int(pixel) for pixel in pix.split(' ')]
    img = np.asarray(img).reshape(48, 48)
    Arr.append(img.astype('float32') / 255.0)
```

```python
X = np.array(Arr)
X = np.expand_dims(X, -1)
```

```python
X.shape
```

```python
y = new_data['emotion'].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state = 42)
```

```python
from keras.layers import Flatten, Dense
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator , img_to_array, load_img
from keras.losses import categorical_crossentropy
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())

model.add(Dense(64, activation='relu'))

model.add(Dense(7, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"\nTest Accuracy: {test_acc}")
```

Data Augmentation for Training data

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
train_generator = datagen.flow(x_train, y_train, batch_size=32)
```

```python
model.fit(train_generator, epochs=2)
```

```python
data_labels = {0:'Angry',1:'Disgust',2:'Afraid',3:'Ecstatic',4:'NO REACTION!',5:'Unhappy',6:'Surprised'}
```

```python
predictions = model.predict(x_test)

for i in range(20):
    predicted_label = np.argmax(predictions[i])
    true_label = y_test[i]
    img = x_test[i]
    plt.title(f'Predicted Label :{predicted_label} ({data_labels[predicted_label]})')
    plt.suptitle(f'Actual Label :{true_label} ({data_labels[true_label]})')
    plt.imshow(img, cmap='gray')
    plt.show()
    print(f"Image {i + 1} - Pred: {predicted_label}, Actual: {true_label}")
```
